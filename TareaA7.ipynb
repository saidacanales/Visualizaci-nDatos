```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df
```

# Generate histograms of the blood pressure variables `sistol` and `diastol` and have them displayed side by side

```python
df.columns
```

```python

sn.displot(data = df.loc[:,
                          ['sistol']
                        ]
           )
plt.xticks(fontsize = 10, rotation = 45)
plt.yticks(fontsize = 10, rotation = 45)
plt.show()

sn.displot(data = df.loc[:,
                          ['diastol']
                        ]
           )
plt.xticks(fontsize = 10, rotation = 45)
plt.yticks(fontsize = 10, rotation = 45)
plt.show()
```

Conclusión: el histograma de distol es asimétrico y el de diastol también es asimétrico. Sí hay datos atípicos en las dos gráficas y se ve claramente por las barras que son disparadas hacia arriba.

# Generate the histograms for the variables `weight` and `waist` and have them displayed contiguously.

```python

sn.displot(data = df.loc[:,
                          ['peso']
                        ]
           )
plt.xticks(fontsize = 10, rotation = 45)
plt.yticks(fontsize = 10, rotation = 45)
plt.show()

sn.displot(data = df.loc[:,
                          ['cintura']
                        ], bins = 15
           )
plt.xticks(fontsize = 10, rotation = 45)
plt.yticks(fontsize = 10, rotation = 45)
plt.show()
```

Conclusion: el histograma de peso está sesgado hacia la derecha y el de la cintura es asimétrico. No hay datos atípicos, todos están muy uniformes

# Generate a scatterplot to analyze the relationship between the variables `waist` and `weight`

```python
plt.scatter(x = 'peso',
               y= 'cintura',
               data = df)
plt.title('Peso vs edad', fontsize = 17, color = 'blue')
plt.xlabel('Peso', fontweight = 'regular', color = 'skyblue', fontsize = 17, horizontalalignment = 'center')
plt.ylabel('Cintura', fontweight = 'regular', color = 'skyblue', fontsize = 17, horizontalalignment = 'center')
plt.show()
```

Conclusión: sí hay datos atípicos en esta gráfica ya que no están centrados con el resto de los datos. 

Cuando el peso aumenta, la cintura también aumenta y vice versa
