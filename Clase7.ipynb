```python
df = pd.read_excel('mic2.xlsx',
                   engine = 'openpyxl')
df.head()
```

```python
def custom_cls(x):
    if (0<= x <=20):
        return '<=20'
    elif (21<= x <= 30):
        return '21 - 30'
    elif (31<= x <= 40):
        return '31 - 40'
    else:
        return '>=40'
```

```python
df['AgeGP'] = df.apply(lambda tbl: custom_cls(tbl['Age']), axis=1)
df.head(2)
```

```python
counts = df['AgeGP'].value_counts()
counts
```

```python
counts = counts.to_frame()
type(counts)
```

```python
counts
```

```python
counts['%'] = counts.iloc[:,:]/counts.loc[:,'AgeGP'].sum(axis=0)*100
counts
```

```python
counts = counts.reindex(['<=20', '21 - 30', '31 - 40', '>=40'])
counts
```

```python
counts = counts.reset_index()
counts
```

```python
counts = counts.rename({'index':'Age_Group', 'AgeGP': 'Age'},
                        axis=1)
counts
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = list(counts['Age_Group'])
y = list(counts['Age'])

plt.subplot(1,2,1)
plt.rcParams["figure.figsize"] = (10,7)
plt.bar(x,y, width = 0.7, color = 'darkcyan')

plt.xlabel('Age group', fontweight = 'regular', color = 'red', fontsize = 14, horizontalalignment = 'center')
plt.ylabel('Counts', fontweight = 'regular', color = 'red', fontsize = 14, horizontalalignment = 'center')

plt.xticks(fontsize = 14, color = 'red')
plt.yticks(fontsize = 14, color = 'red')
plt.title('Number of respondents and their age group', fontsize = 13, color = 'red')

plt.subplot(1,2,2)
color_list = ['gold', 'yellowgreen', 'coral', 'skyblue', 'pink']
explode_list = [0.02, 0.04, 0.07, 0.09]
counts['Age'].plot(kind = 'pie',
                  figsize = (15,6),
                  labels = None,
                   autopct = '%1.1f%%',
                   pctdistance = 1.2,
                   shadow = True,
                   startangle = 90,
                   colors = color_list,
                   explode = explode_list
                  )
```
