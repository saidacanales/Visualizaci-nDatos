```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# Description of Dataset

With this dataset we will analyze the information regarding the deaths and causes of deaths in Mexico City in the year 2021.

```python
df = pd.read_csv('defunciones_corte.csv'
                  )
df = df.iloc[:10000]
df.head(2)
```

```python
df.columns
```

# Description of the Variables 

`edad` is the age of the person was when they died. `sexo` is the gender of the person who died. `fec_defuncion` is the date of the person when they died. `estado` is the state the person was in when they died, they being Estado de Mexico or Ciudad de Mexico. `causa` is either covid or other and explained further in the next column. `causa_registro` is the cause registered for the person's death. `alcaldia` is the colony of the person's origin. `LugarMuerte` is the place the person was when they died that being home or hospital. `num_consecutivo` is the number they were on the death list.

# 1. What is the average age for deaths in Mexico City?

```python
sns.boxplot(data = df.loc[:,
                          ['edad']
                        ] 
           )
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17, rotation = 45)
plt.title('Age of death', fontsize = 17)
plt.show()
```

In this graph, we can see that the average age of death in Mexico City is 70, with an average minimum of aroung 58 and average maximum of 80.

# 2. What is the relationship between the age of the people who died and their gender?

```python
sns.stripplot(x = 'sexo',
               y= 'edad',
               data = df)
plt.title('Age of death vs gender', fontsize = 17)
plt.show()
```

With this graph, we can see the deaths per gender for both men and women in Mexico City. It is interesting to see how women tend to live longer, since the more concentratino of points is in the later years of age. Also, note how fewer women die in their teens and early 20s compared to men.

# 4. What is the average of people who died because of ABDOMEN AGUDO, DIABETES MELLITUS TIPO 2, HIPERTENSION ARTERIAL.

```python
myPivot = pd.pivot_table(df,
                         aggfunc = ['mean'],
                         index = ['causa_registro'],
                         margins = True,
                         columns = ['causa_registro'],
                        )
myPivot
```

Here, we can see the average number of people who died because of ABDOMEN AGUDO, DIABETES MELLITUS TIPO 2, HIPERTENSION ARTERIAL is 86. We can also see the average number of people who died because of the other causes of death.
