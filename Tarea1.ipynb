# Antromometria-dataset

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df
```

```python
df.columns
```

```python
df.drop(['folio', 'intp', 'entidad', 'edad', 'meses',
       'ropa', 'talla', 'emb', 'temb', 'cintura', 'cadera', 'sistol',
       'diastol', 'hpresion', 'tbrazo', 'htension', 'PrimaryLast', 'code_upm',
       'est_dis', 'est_urb', 'est_marg', 'pondef', 'est_var'],
       axis = 1,
        inplace=True
       )
df.columns
```

```python
df.rename(columns = {'desc_ent':'State', 'sexo':'Gender', 'peso':'Weight',},
        inplace = True)
df.head()
```

```python
df.set_index('State', inplace=True)
df
```

```python
df.reset_index()
myPivot = pd.pivot_table(df,
                         index=['State'],
                         values = ['Weight'],
                         aggfunc = ['mean'],
                         columns = ['Gender'],
                         margins = True
                        )
myPivot
```
